```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd

# Inisialisasi analyzer
analyzer = SentimentIntensityAnalyzer()

# Fungsi untuk menganalisis sentimen
def analyze_sentiment(text):
    sentiment_score = analyzer.polarity_scores(text)
    compound_score = sentiment_score['compound']  # Skor komposit untuk sentimen keseluruhan

    # Menentukan label berdasarkan skor komposit
    if compound_score >= 0.05:
        sentiment_label = 'Positive'
    elif compound_score <= -0.05:
        sentiment_label = 'Negative'
    else:
        sentiment_label = 'Neutral'

    return compound_score, sentiment_label

# Fungsi untuk pra-pemrosesan teks (sesuaikan dengan kebutuhan Anda)
def preprocess_text(text):
    # Tempatkan kode pra-pemrosesan di sini jika perlu
    return text

# Terapkan pra-pemrosesan dan analisis sentimen ke kolom 'full_text'
df['cleaned_text'] = df['full_text'].apply(preprocess_text)  # Pra-pemrosesan teks

# Terapkan analisis sentimen dan pisahkan hasilnya menjadi 2 kolom: skor dan label
df[['sentiment_score', 'sentiment']] = df['cleaned_text'].apply(lambda x: pd.Series(analyze_sentiment(x)))

# Menampilkan hasil dengan kolom tambahan
print(df[['full_text', 'sentiment_score', 'sentiment']].head())
```

```python
df = df[['Date', 'sentiment_score']]
df
```

```python
# Konversi kolom Date menjadi datetime
df['Date'] = pd.to_datetime(df['Date'])

# Agregasi: Rata-rata skor sentimen per hari
df_daily_sentiment = df.groupby('Date')['sentiment_score'].mean().reset_index()

# Membatasi skor sentimen menjadi dua angka di belakang koma
df_daily_sentiment['sentiment_score'] = df_daily_sentiment['sentiment_score'].round(2)

print(df_daily_sentiment)
```
