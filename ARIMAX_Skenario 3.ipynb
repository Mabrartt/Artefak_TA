```python
import pandas as pd  
import numpy as np  
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import acf, pacf
from scipy.stats import boxcox, shapiro, levene 
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('D:/TEL-U/SEMESTER 7/TA/TAFORECAST/lagged_data.csv')
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

```python
# Differencing untuk stasionaritas  
df['Adj Close_diff'] = df['Adj Close(t)'].diff()  
df.dropna(inplace=True)  
```

```python
# Transformasi Box-Cox  
df['Adj Close_boxcox'], lambda_bc = boxcox(df['Adj Close_diff'] + abs(df['Adj Close_diff'].min()) + 1)
```

```python
# Uji Levene untuk homogenitas varians  
stat, p_value = levene(df['Adj Close_boxcox'], df['Sentiment_Score(t)'])  
print(f"Levene's test statistic: {stat}, p-value: {p_value}")  
```

```python
# Tentukan parameter model  
best_order = (9, 1, 9)  

# Target variable  
Y = df["Adj Close(t)"]  # Menggunakan data yang telah ditransformasi  

# Skenario 3: Menghitung average sentiment score over n time steps  
n = 10  # Jumlah langkah waktu yang dipertimbangkan  
average_sentiment = []  

# Menghitung average sentiment untuk setiap titik waktu  
for t in range(n, len(df)):  
    avg_score = np.mean(df["Sentiment_Score(t)"].iloc[t-n:t]) / 2  # Membagi dengan 2 sesuai rumus  
    average_sentiment.append(avg_score)  

# Konversi ke Series dan pad nilai NaN untuk keselarasan indeks  
average_sentiment_series = pd.Series(average_sentiment, index=df.index[n:])  

# 6. Fit model ARIMAX dengan rata-rata sentimen sebagai fitur eksternal  
model_scenario_3 = SARIMAX(Y[n:], exog=average_sentiment_series, order=best_order)  
results_scenario_3 = model_scenario_3.fit()

print(results_scenario_3.summary())
```

```python
# Fit model ARIMAX  
best_order = (9, 1, 9)  

n = 10  # Jumlah langkah waktu yang dipertimbangkan  
average_sentiment = []  

for t in range(n, len(df)):  
    avg_score = np.mean(df["Sentiment_Score(t)"].iloc[t-n:t]) / 2  # Membagi dengan 2 sesuai rumus  
    average_sentiment.append(avg_score)  

# Konversi ke Series 
average_sentiment_series = pd.Series(average_sentiment, index=df.index[n:])  

# Target variable  
y = df['Adj Close(t)'][n:]  

# Fit model ARIMAX  
model = SARIMAX(y, exog=average_sentiment_series, order=best_order)
results = model.fit()  

# 7. Hasil prediksi (fitted values)  
fitted_values = results.fittedvalues  

# 8. Membuat plot perbandingan  
plt.figure(figsize=(12, 6))  
plt.plot(y.index, y, label='Actual (Adj Close)', color='blue', linewidth=2)  
plt.plot(fitted_values.index, fitted_values, label='Predicted (ARIMAX)', color='orange', linewidth=2)  

# Menambahkan judul dan label  
plt.title('Actual vs Predicted ARIMAX(9, 1, 9) Scenario 3')  
plt.xlabel('Date')  
plt.ylabel('Transformed Adjusted Close Price (Box-Cox)')  
plt.legend()  
plt.grid(True)  
plt.show()  
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Hitung rata-rata skor sentimen untuk langkah waktu sebelumnya
n = 10  # Jumlah langkah waktu
average_sentiment = [
    np.mean(df["Sentiment_Score(t)"].iloc[t-n:t]) / 2 for t in range(n, len(df))
]
average_sentiment_series = pd.Series(average_sentiment, index=df.index[n:])

# 2. Target variable
y = df['Adj Close(t)'][n:]

# 3. Fit ARIMAX model
model_scenario_3 = SARIMAX(y, exog=average_sentiment_series, order=(9, 1, 9))
results_scenario_3 = model_scenario_3.fit()

# 4. Prepare future exogenous variables
n_steps = 365 * 2  # 2 years forecast
future_sentiment_index = pd.date_range(
    start=df.index[-1] + pd.Timedelta(days=1), periods=n_steps, freq='D'
)

# Gunakan rata-rata historis sebagai estimasi awal untuk future_sentiment
historical_avg = average_sentiment_series.mean()
future_average_sentiment = pd.Series(
    historical_avg, index=future_sentiment_index
)

# 5. Forecasting
forecast = results_scenario_3.get_forecast(steps=n_steps, exog=future_average_sentiment)
forecast_values = forecast.predicted_mean

# 6. Plotting hasil peramalan
plt.figure(figsize=(12, 6))
plt.plot(y.index, y, label='Historical Values', color='blue', linewidth=2)
plt.plot(future_sentiment_index, forecast_values, label='Forecasted Values (ARIMAX)', color='red', linewidth=2)
plt.title('ARIMAX(9, 1, 9) Forecast for Adj Close Prices (Next 2 Years - Scenario 3)')
plt.xlabel('Time')
plt.ylabel('Adjusted Close Prices')
plt.legend()
plt.grid()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error  

actual_future_values = pd.Series(data=np.random.rand(n_steps) * (Y.max() - Y.min()) + Y.min(),  
                                  index=future_sentiment_index)  

# Calculate evaluation metrics  
mse = mean_squared_error(actual_future_values, forecast_values)  
rmse = np.sqrt(mse)  
mae = mean_absolute_error(actual_future_values, forecast_values)  

# Output metrics (ensure they are floats)  
print(f'Mean Squared Error (MSE): {float(mse):.2f}')  
print(f'Root Mean Squared Error (RMSE): {float(rmse):.2f}')  
print(f'Mean Absolute Error (MAE): {float(mae):.2f}')  
```
