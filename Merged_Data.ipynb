```python
import pandas as pd
```

```python
df = pd.read_csv('/content/bitcoin_historical_data (1).csv')
df
```

```python
# Fungsi untuk mengekstrak tanggal tanpa waktu
def extract_date(date_str):
    return pd.to_datetime(date_str).strftime('%Y-%m-%d')

# Terapkan fungsi ke kolom Date
df['Date'] = df['Date'].apply(extract_date)
```

```python
df
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df1 = pd.read_csv('/content/Dataset_cleaned_final2.csv')
df1
```

```python
df1["Date"] = pd.to_datetime(df1["Date"])

# Membuat range tanggal penuh
date_range = pd.date_range(start=df1["Date"].min(), end=df1["Date"].max())

# Membuat DataFrame dengan tanggal lengkap
full_df = pd.DataFrame({"Date": date_range})

# Gabungkan dengan data asli
df1 = full_df.merge(df1, on="Date", how="left")

# Interpolasi nilai yang hilang
df1["sentiment_score"] = df1["sentiment_score"].interpolate()
```

```python
df1
```

```python
# Gabungkan dataset berdasarkan kolom 'Date'
data = pd.merge(df, df1, on="Date", how="inner")
```

```python
data
```

```python
# Filter data from January 1, 2020
filtered_data = data[data["Date"] >= "2020-01-01"].reset_index(drop=True)

# Function to create lagging features
def create_lags(df, lags):
    lagged_data = df.copy()
    for lag in range(1, lags + 1):
        lagged_data[f"Close(t-{lag})"] = lagged_data["Close"].shift(lag)
        lagged_data[f"Open(t-{lag})"] = lagged_data["Open"].shift(lag)
        lagged_data[f"High(t-{lag})"] = lagged_data["High"].shift(lag)
        lagged_data[f"Low(t-{lag})"] = lagged_data["Low"].shift(lag)
        lagged_data[f"Sentiment_Score(t-{lag})"] = lagged_data["sentiment_score"].shift(lag)
        lagged_data[f"Adj Close(t-{lag})"] = lagged_data["Adj Close"].shift(lag)

    # Current t column
    lagged_data = lagged_data.rename(columns={
        "Close": "Close(t)",
        "Open": "Open(t)",
        "High": "High(t)",
        "Low": "Low(t)",
        "sentiment_score": "Sentiment_Score(t)",
        "Adj Close": "Adj Close(t)"
    })

    # Drop rows with NaN values due to lagging
    lagged_data = lagged_data.dropna().reset_index(drop=True)

    return lagged_data

# Create lags up to t-10
lagged_data = create_lags(filtered_data, lags=10)
```

```python
# Reorder columns as requested
columns_order = ["Date"]
for lag in range(10, 0, -1):
    columns_order.extend([
        f"Close(t-{lag})", f"Open(t-{lag})", f"High(t-{lag})", f"Low(t-{lag})", f"Sentiment_Score(t-{lag})", f"Adj Close(t-{lag})"
    ])
columns_order.extend(["Close(t)", "Open(t)", "High(t)", "Low(t)", "Sentiment_Score(t)", "Adj Close(t)"])

# Reorder the DataFrame
lagged_data = lagged_data[columns_order]

print(lagged_data.head())
```
